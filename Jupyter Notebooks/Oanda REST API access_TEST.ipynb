```python
import requests
import configparser
# import oandapyV20
# import oandapyV20.endpoints.accounts as accounts
# import oandapyV20.endpoints.instruments as instruments

# Load the credentials from oanda.cfg
config = configparser.ConfigParser()
config.read('H:/My Drive/00 PROJECTS/0.3 Coding/Jupyter Notebooks/oanda.cfg')

# Extract details from the configuration file
OANDA_ACCOUNT_ID = config['oanda']['account_id']
OANDA_API_KEY = config['oanda']['access_token']
OANDA_ACCOUNT_TYPE = config['oanda']['account_type']

# Initialize the OANDA API client
# client = oandapyV20.API(access_token=OANDA_ACCESS_TOKEN)

OANDA_API_URL = f"https://api-fxpractice.oanda.com/v3/accounts/{OANDA_ACCOUNT_ID}/instruments"

# Set the headers for authentication
headers = {
    "Authorization": f"Bearer {OANDA_API_KEY}",
    "Content-Type": "application/json"
}

# Fetch the list of instruments from the OANDA REST API
def get_currency_symbols():
    try:
        # Make the request to OANDA's AccountInstruments endpoint
        response = requests.get(OANDA_API_URL, headers=headers)
        response.raise_for_status()  # Check for errors
        
        # Extract the instruments data
        instruments = response.json()['instruments']
        
        # Filter only instruments that are of type 'CURRENCY'
        currency_symbols = [instrument for instrument in instruments if instrument['type'] == 'CURRENCY']
        return currency_symbols
    except requests.exceptions.RequestException as e:
        print(f"Error fetching instruments: {e}")
        return None

# Fetch and print the CURRENCY symbols
currency_symbols = get_currency_symbols()

if currency_symbols:
    print("List of available CURRENCY symbols:")
    for instrument in currency_symbols:
        print(f"Symbol: {instrument['name']}, Type: {instrument['type']}")
else:
    print("Failed to retrieve currency symbols.")
```
